# 성별로 분류 후, 연령대별 비만 환자 수 그래프 확인

```python
import csv
import matplotlib.pyplot as plt
from collections import defaultdict

# 파일 열기
f = open('gender2023.csv', encoding='cp949')  # 파일 경로 사용
data = csv.reader(f)

# 데이터 처리
next(data)  # 헤더 건너뛰기

# 연령대별 여성 환자 수 집계
age_groups = defaultdict(int)

for row in data:
    if row[3].strip() == '남':  # 성별이 정확히 "남"인 경우만
        age_groups[row[4]] += int(row[5])  # 연령대별로 환자 수 합산

f.close()

# 가장 환자 수가 많은 연령대 찾기
max_age_group = max(age_groups, key=age_groups.get)
max_value = age_groups[max_age_group]

print(f"가장 환자 수가 많은 연령대는 {max_age_group}이며, 환자 수는 {max_value}명입니다.")

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(age_groups.keys(), age_groups.values(), color='skyblue')
plt.title('성별: 남자 - 연령대별 환자 수')
plt.xlabel('Age Group')
plt.ylabel('Patients')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 두 결과 비교시 남자가 더 비만도가 높은지 여자가 더 비만도가 높은지 원형 그래프

```python
import csv
import matplotlib.pyplot as plt

# 파일 열기
f = open('gender2023.csv', encoding='cp949')  # 파일 경로
data = csv.reader(f)
next(data)  # 헤더 건너뛰기

# 남성과 여성의 환자수 합계를 저장할 변수
male_total = 0
female_total = 0

# 데이터 필터링
for row in data:
    if '남' in row[3]:  # 성별이 '남'인 경우
        male_total += int(row[5])  # 환자수 누적
    elif '여' in row[3]:  # 성별이 '여'인 경우
        female_total += int(row[5])  # 환자수 누적

f.close()

# 결과 출력
print(f"남성 총 환자 수: {male_total}")
print(f"여성 총 환자 수: {female_total}")

# 원형 그래프 데이터
labels = ['남성', '여성']
sizes = [male_total, female_total]
colors = ['blue', 'pink']
explode = (0.1, 0)  # 남성 원형을 약간 분리하여 강조

# 원형 그래프 그리기
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode)
plt.title('남성과 여성의 비만 환자 비율')
plt.axis('equal')  # 원형 유지
plt.show()
```

# 지정한 연령대별 남성과 여성 환자 수 비교 막대 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# FutureWarning 숨기기
warnings.simplefilter(action='ignore', category=FutureWarning)


# 파일 읽기
data_csv = pd.read_csv('gender2023.csv', encoding='euc-kr')

# 특정 연령대를 입력받아 성별 환자 수 비율 확인
def plot_gender_ratio_by_age(age_group):
    # 입력된 연령군 데이터 필터링
    filtered_data = data_csv[data_csv['연령군'] == age_group]
    
    if filtered_data.empty:
        print(f"'{age_group}'에 해당하는 데이터가 없습니다. 올바른 연령대를 입력하세요.")
        return
    
    # 성별로 그룹화하여 환자 수 계산
    gender_summary = filtered_data.groupby('성별')['환자수'].sum().reset_index()
    
    # 남성, 여성 환자 수 계산
    male_total = gender_summary.loc[gender_summary['성별'] == '남', '환자수'].values[0] if '남' in gender_summary['성별'].values else 0
    female_total = gender_summary.loc[gender_summary['성별'] == '여', '환자수'].values[0] if '여' in gender_summary['성별'].values else 0

    # 결과 출력
    print(f"연령대 '{age_group}'의 남성과 여성 환자 수:")
    print(f"- 남성 총 환자 수: {male_total}")
    print(f"- 여성 총 환자 수: {female_total}")
    
    # 시각화
    plt.figure(figsize=(8, 6))
    sns.barplot(data=gender_summary, x='성별', y='환자수', palette='Set2')
    plt.title(f"연령대 '{age_group}'의 남성과 여성 환자 수 비교", fontsize=16)
    plt.xlabel('성별', fontsize=14)
    plt.ylabel('환자수', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    plt.show()

# 사용 예시: 특정 연령대 입력
age_group = input("확인할 연령대를 입력하세요 (예: '30~39세'): ")
plot_gender_ratio_by_age(age_group)
```

# 본인이 지정한 성별, 연령대별 환자 수 점 그래프

```python
import warnings
import csv
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import pandas as pd

# FutureWarning 숨기기
warnings.simplefilter(action='ignore', category=FutureWarning)

# 파일 열기
f = open('gender2023.csv', encoding='cp949')  # 파일 경로 사용
data = csv.reader(f)

# 데이터 처리
next(data)  # 헤더 건너뛰기

# 연령대별 환자 수 집계 (성별 구분)
age_groups = defaultdict(lambda: {'남': 0, '여': 0})

for row in data:
    if row[3].strip() == '여':  # 성별이 "여"인 경우
        age_groups[row[4]]['여'] += int(row[5])  # 해당 연령대에 여성 환자수 합산
    elif row[3].strip() == '남':  # 성별이 "남"인 경우
        age_groups[row[4]]['남'] += int(row[5])  # 해당 연령대에 남성 환자수 합산

f.close()

# 사용자로부터 연령대와 성별 입력 받기
selected_age_groups = input("점 그래프를 그릴 연령대를 입력하세요 (예: 10~19세, 20~29세 등): ").split(',')
selected_gender = input("성별을 입력하세요 (남, 여): ").strip()

# 입력된 연령대와 성별만 필터링
filtered_data = []

for age_group, genders in age_groups.items():
    if age_group in selected_age_groups:
        if selected_gender == '남':
            filtered_data.append({'연령군': age_group, '성별': '남', '환자수': genders['남']})
        elif selected_gender == '여':
            filtered_data.append({'연령군': age_group, '성별': '여', '환자수': genders['여']})

# 필터링된 데이터를 DataFrame 형식으로 변환
data_for_plot = pd.DataFrame(filtered_data)

# 선택된 연령대와 성별에 대한 환자 수 출력
if len(filtered_data) > 0:
    print(f"선택한 연령대: {', '.join(selected_age_groups)}")
    print(f"선택한 성별: {selected_gender}")
    print("선택된 연령대와 성별의 환자 수:")
    print(data_for_plot[['연령군', '성별', '환자수']])
else:
    print("선택한 연령대와 성별에 해당하는 데이터가 없습니다.")

# 점 그래프 그리기
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_for_plot, x='연령군', y='환자수', hue='성별', style='성별', palette='Set2', s=100)

# 제목 및 레이블 설정
plt.title(f'선택한 연령대 및 성별의 환자 수 (점 그래프)', fontsize=16)
plt.xlabel('연령대', fontsize=14)
plt.ylabel('환자수', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
